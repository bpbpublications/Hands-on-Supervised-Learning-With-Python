```python
cols = ['Name','Sex', 'Ticket','Fare','Cabin','Embarked']
```

```python
df = df.drop(columns = ['Cabin', 'Name', 'PassengerId', 'Ticket', 'SibSp'])
```

```python
df.fillna(method='ffill', inplace=True)
df.fillna(method='ffill', inplace=True)
```

```python
lab_enc = preprocessing.LabelEncoder()
df['Sex'] = lab_enc.fit_transform(df['Sex'])
df['Fare'] = lab_enc.fit_transform(df['Fare'])
df['Embarked'] = lab_enc.fit_transform(df['Embarked'])
```

```python
corr = df.corr()
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr,cmap=cmap, vmax=.3,square=True,linewidths=6, cbar_kws={"shrink": .5})
colormap = plt.cm.viridis
```

```python
plt.figure(figsize=(12,12))
plt.title('Pearson Correlation of Features', y=1.05, size=15)
sns.heatmap(df.corr(),linewidths=0.1,vmax=1.0, square=True, cmap=colormap, linecolor='white',
annot=True)
```

```python
Y = df['Survived']
X = df.drop(columns = ['Survived'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
